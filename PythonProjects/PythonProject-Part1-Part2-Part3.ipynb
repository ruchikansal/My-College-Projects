# Part II: Data Preparation and Cleaning

### 8. Rename the columns belonging to Products by removing Mnt from them.

```python
df.rename(columns={'MntWines': 'Wines',
                     'MntFruits': 'Fruits',
                     'MntMeatProducts': 'Meat',
                     'MntFishProducts': 'Fish',
                     'MntSweetProducts': 'Sweet',
                     'MntGoldProds': 'Gold'}, inplace=True)

df.columns
```

### 9. Rename the columnsbelonging to Place by removing Num and Purchases from them

```python
df.rename(columns={'NumWebPurchases': 'Web',
                     'NumCatalogPurchases': 'Catalog',
                     'NumStorePurchases': 'Store'}, inplace=True)

df.columns
```

### 10. Keep only NumDealsPurchases column in the Promotion section

```python
columns_to_delete = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']
df.drop(columns=columns_to_delete, inplace=True)
df
```

### 11. Delete the column NumWebVisitsMonth

```python
df.drop(columns='NumWebVisitsMonth', inplace=True)
df
```

### 12. The attribute Year_Birth is not helpful to segment your data. Instead create a new derived column named Age.

```python
from datetime import datetime
current_year = datetime.now().year

df['Age'] = current_year - df['Year_Birth']
df
```

### 13. Create a new column named Spending that holds the total amount spent on all products categories.

```python
df['Spending'] = df[['Wines', 'Fruits', 'Meat', 'Fish', 'Sweet', 'Gold']].sum(axis=1)
df
```

### 14. Change the Dt_Customer type from object to datetime. Use the format YYYY-MM-DD

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y').dt.strftime('%Y-%m-%d')
df
# '.dt' is used to access the datetime components of the 'Dt_Customer' column 
# after converting it to a datetime format using pd.to_datetime(). 
# In this case, .dt.strftime('%Y-%m-%d') is used to format the datetime values as strings in the "yyyy-mm-dd" format.
```

### 15. Show the unique values of the column data Marital_Status

```python
df['Marital_Status'].unique()
```

### 16. Change the column Marital status

```python
df['Marital_Status'].replace(['Divorced', 'Single', 'Absurd', 'Widow','YOLO'], 'Alone', inplace=True)

df['Marital_Status'].replace(['Married', 'Together'], 'Not Alone', inplace=True)

df
```

```python
df['Marital_Status'].unique()
```

### 17. Show the unique values of the column Education

```python
df['Education'].unique()
```

### 18. Change the column Education

```python
df['Education'].replace(['Basic', '2n Cycle'], 'Undergraduate', inplace=True)
df['Education'].replace(['Graduation', 'PhD', 'Master'], 'Postgraduate', inplace=True)
df
df['Education'].unique()
```

### 19. Have a look at column Income. Investigate the presence of outliers? Delete them accordingly

### Boxplot before deleting the outliers

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(8, 6)) 
df.boxplot(column='Income')
plt.title('Boxplot of Income')
plt.ylabel('Income')
plt.grid(False)
plt.show()
```

```python
mean_income = df['Income'].mean()
std_income = df['Income'].std()
#outliers = df[(df['Income'] > (mean_income + 3 * std_income)) | (df['Income'] < (mean_income - 3 * std_income))]
#df.drop(outliers.index, inplace=True)
df.loc[(df['Income'] > (mean_income + 3 * std_income)) | (df['Income'] < (mean_income - 3 * std_income)), 'Income'] = np.nan
df.shape
```

#### we are using the standard deviation to identify outliers in the 'Income' column. By calculating the standard deviation, we can determine a threshold beyond which values are considered outliers. In this case, outliers are values that lie more than 3 standard deviations away from the mean income value. There were total 8 outliers in this income column.

### Boxplot after deleting the outliers

```python
plt.figure(figsize=(8, 6)) 
df.boxplot(column='Income')
plt.title('Boxplot of Income')
plt.ylabel('Income')
plt.grid(False)
plt.show()
```

### 24. Create a new column Has_Complaint

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
two_years_ago = pd.Timestamp.now() - pd.DateOffset(years=2)

# Create 'Has_Complaint' column based on conditions
df['Has_Complaint'] = df.apply(lambda row: 'Has complaint' if row['Complain'] == 1 and row['Dt_Customer'] >= two_years_ago else 'No complaint', axis=1)
df

#df[df['Has_Complaint'] == 'No complaint']
# There are no customers who has complained in the last 2 years
```
